```python
cars=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
```

```python
cars.sort_values(by='mpg')
```

```python
cars.iloc[range(11)] #first 11 cars
```

```python
cars.head(11)   ## first 11 cars
cars.tail(11)   ## last 11 cars
```

```python
cars.sort_values(by='mpg').head()
```

```python
cars[['mpg', 'cylinders','name']].head()
```

```python
cars.sort_values(by='horsepower' ,ascending=False).head()
```

```python
## Show the cars that has an mpg lower than 15

cars['mpg']>15


cars[cars['mpg']>15]  #cars that have an mpg of greater than 15
```

```python
cars[cars['horsepower']>150]   ## order this list in a descending order based on their mpg
```

```python
super_cars=cars[cars['horsepower']>150]

super_cars.sort_values(by='mpg', ascending=False)
```

```python
cars[cars['horsepower']>150].sort_values(by='mpg', ascending=False)
```

```python
sorted_cars=cars.sort_values(by='mpg', ascending=False)

sorted_cars[sorted_cars['horsepower']>150]
```

```python
cars['horsepower']>150 #criterion 1
```

```python
cars['mpg']>15    #criterion 2 
```

```python
(cars['horsepower']>150)&(cars['mpg']>15)   ## Shows the cars that meets cretrion 1 and 2
```

```python
cars[(cars['horsepower']>150)&(cars['mpg']>15) ]    ##   & means AND 
```

```python
 cars[(cars['horsepower']>150)|(cars['mpg']>15) ]   ##   | means OR 
```
