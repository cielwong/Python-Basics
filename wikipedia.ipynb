```python
summaries=dict()
for  a_language in all_languages:
    try:
        lan_summary=wikipedia.page(a_language).summary
        summaries[a_language]=lan_summary
    except:
        continue
```

```python
page_title='Machine Learning'
start_date='20190101'
end_date='20200101'
url='https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/{}/monthly/{}/{}'.format(page_title, start_date, end_date)
```

```python
start_date='20190101'
end_date='20200101'
all_rs=[]
for a_language in all_languages:
    page_title=a_language.replace(' ', '%20')
    try: 
        url=url='https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/{}/monthly/{}/{}'.format(page_title, start_date, end_date)
        results=urllib.request.urlopen(url).read()
    except:
        print(page_title, 'not found')
        continue
    rs=json.loads(results)['items']
    #print(rs)
    views=pd.DataFrame(rs)[['timestamp','views']]
    views['lang']=a_language
    all_rs.append(views)
```

```python
final=pd.concat(all_rs)
```

```python
final['timestamp']=final['timestamp'].str.rsplit('00', 1, expand=True)[0]
```

```python
final['timestamp']=pd.to_datetime(final['timestamp'])
```

```python
import seaborn as sns
```

```python
final.sort_values(by='timestamp')
```

```python
sns.relplot(x='timestamp', y='views', kind='line', data=final )
```

```python
sns.relplot(x='timestamp', y='views', hue='lang', kind='line', data=final )
```

```python
top_ten=final.groupby('lang')['views'].agg('mean').reset_index().sort_values(by='views', ascending=False)[:10]
```

```python
top_ten_names=top_ten['lang']
```

```python
final_ten=final[final['lang'].isin(top_ten_names)]
```

```python
sns.relplot(x='timestamp', y='views', hue='lang', kind='line', data=final_ten )
```
