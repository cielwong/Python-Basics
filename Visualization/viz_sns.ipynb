```python
# Import necessarily libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
car_ads=pd.read_csv('classified_ads_for_cars_originial.csv')



sns.set(style="darkgrid")
```

## Emphasizing continuity with lines

```python
sns.relplot(x='manufacture_year', y='mean', kind='line',data=yearly_data)
```

```python
sns.relplot(x='manufacture_year', y='price_eur', data=car_ads)
```

```python
sns.relplot(x='manufacture_year', y='price_eur', kind='line',data=car_ads)
```

```python
sns.relplot(x='manufacture_year', y='price_eur', kind='line', ci=None , data=car_ads)
```

```python
sns.relplot(x='manufacture_year', y='price_eur', kind='line', ci="sd" , data=car_ads)
```

```python
sns.relplot(x='manufacture_year', y='price_eur', hue='transmission' ,kind='line', ci="sd" , data=car_ads)
```

##  Categorical data

```python
bmw=car_ads[car_ads['maker']=='bmw']
```

```python
sns.catplot(y='price_eur', x='seat_count', data=bmw)
```

```python
sns.catplot(y='price_eur', x='seat_count', jitter=False,  data=bmw)
```

```python
sns.catplot(y='price_eur', x='seat_count', kind='swarm',  data=bmw)
```

```python
sns.catplot(y='price_eur', x='seat_count', hue='fuel_type',kind='swarm',  data=bmw)
```

```python
sns.catplot(x="fuel_type", y="price_eur", hue="transmission" , kind="bar", data=car_ads);
```

```python
#If not by median
```

```python
medians=car_ads.groupby(['fuel_type','transmission'])['price_eur'].agg(['median']).reset_index()
```
